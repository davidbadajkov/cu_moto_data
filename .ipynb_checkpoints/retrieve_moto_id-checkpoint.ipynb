```python
import requests
from bs4 import BeautifulSoup
import json
import time
import math
import pandas as pd
from datetime import date
import numpy as np
import pickle

from IdScraper import extract_nb_moto
from IdScraper import create_url_list
from IdScraper import extract_moto_id
```

<h2> Create url for each page </h2>
<p> First, extract the number of pages per brand </p>
<p> Here we made a list of all the moto brands you will find on the second-hand market of Belgium. For each brand it is possible to retrieve easily all the moto id. 30 motorcycles are shown for each page for a brand. Some brands have more pages than others as you can see from the list 'amount_pages'. This is retrieved manually. </p>
<p> The logic of the page url is the same for each brand so we easily loop over it. Below is an example of how the page looks like. </p>

```python
list_motorcycles = ['aprilia', 'bmw', 'buell', 'cagiva', 'ducati', 'harley-davidson', 'honda', 'husqvarna', 
                 'hyosung', 'kawasaki', 'ktm', 'mash', 'moto-guzzi', 'mv-agusta', 'piaggio', 'royal-enfield', 
                 'suzuki', 'triumph', 'yamaha']
```

```python
# Extract the number of pages per brand
list_number_pages = []
for moto_brand in list_motorcycles:
    extract = extract_nb_moto(moto_brand)
    x = np.random.randint(low=3,high=5)
    time.sleep(x)
    list_number_pages.append(extract)
```

```python
# Create the necessary URL for further data collection
list_total_url = create_url_list(list_number_pages)
```

```python
# SAVE THE LIST
with open('list_total_url.ob', 'wb') as fp:
    pickle.dump(list_total_url, fp)
```

<h2> Retrieving motorcycles' ID from pages </h2>
<p> In this section we use the previously retrieved url to scrape the moto-id that is on the page. Each page has 30 moto, so we should retrieve 375 page * 30 moto = 11 250 moto id. Some will be duplicate or won't be a moto at all, so we'll have less than that after the preprocessing. </p>

```python
# Opens the file with all the page URL.
with open ('list_total_url.ob', 'rb') as fp:
    list_total_url = pickle.load(fp)
```

```python
'''
import random
random.shuffle(list_total_url)
'''
```

```python
# Extract moto id from pageURL
list_all_id = []
for j in list_total_url:
    x = np.random.randint(low=30,high=50)
    time.sleep(x)
    list_all_id.extend(extract_moto_id(j))
```
